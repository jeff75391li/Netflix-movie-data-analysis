# Data analysis on IMDB scores of actors in Netflix movies/shows

## Goal

In this project, we aim to evaluate and clean the Netflix movies/shows rating dataset and to retrieve the actors with highest rating in each genre.

## Introduction

We have a dataset that includes various information of all Netflix movies/shows available in the United States from year 1945 to 2022. The dataset is represented by two tables `titles.csv` and `credits.csv`.

`titles.csv` describes information of Netflix movies/shows including titles, genres, ratings, whereas `credits.csv` has information of directors and actors that have participated in the movies/shows, including their names.

For each column of `titles.csv`:
- `id`: unique id of the movie/show.
- `title`: movie/show title.
- `type`: type of the show (movie or show).
- `description`: short description of the movie/show.
- `release_year`: the year that the movie/show was released.
- `age_certification`: age certificates for movies/shows.
- `runtime`: runtime of the movie/show.
- `genres`: genre of the movie/show.
- `production_countries`: the countries that the movie/show was produced.
- `seasons`: seasons of the show (not applicable for movies).
- `imdb_id`: IMDB id related to the movie/show.
- `imdb_score`: rating from IMDB.
- `imdb_votes`: number of votes from IMDB.
- `tmdb_popularity`: popularity of the movie/show from TMDB.
- `tmdb_score`: rating from TMDB.

For each column of `credits.csv`:
- `person_ID`: unique id of the director/actor.
- `id`: movie/show id that the person has participated in.
- `name`: name of the person.
- `character`: character name that the person has played in a movie/show.
- `role`: role of the person in a movie/show (directors or actors).

## Data evaluation and cleansing

Here we are going to evaluate the data in `original_titles` and `original_credits` in terms of structure and validity. Then we will perform data cleansing if needed.

We make copies of these tables and assign them `cleaned_titles` and `cleaned_credits` for better readability.

```python
cleaned_titles = original_titles.copy()
cleaned_credits = original_credits.copy()
```

### Evaluate the data structure

First we are going to check `cleaned_titles`.

```python
cleaned_titles.head(10)
```

From the first 10 records in the `cleaned_titles`, we observe that there are multiples values in the column `genres` and `production_countries`. In order to better analyze the data we will have to split the values to ensure one single value per column.

Firstly, let's deal with `genres`. We will look at its value type.

```python
cleaned_titles['genres'][1]
```

Notice that each value in this column is a string instead of an actual list of strings indicating the genres. Therefore we will have to convert it to a list.

```python
cleaned_titles['genres'] = cleaned_titles['genres'].apply(lambda s: eval(s))
cleaned_titles['genres'][1]
```

Then we will use `explode` method in DataFrame to split the values.

```python
cleaned_titles = cleaned_titles.explode("genres")
cleaned_titles.head(10)
```

Problem solved. Now let's look at `production_countries`.

```python
cleaned_titles['production_countries'][1]
```

We notice the same issue as in `genres`. Let's convert them into a list of strings.

```python
cleaned_titles['production_countries'] = cleaned_titles['production_countries'].apply(lambda s: eval(s))
cleaned_titles['production_countries'][1]
```

Let's use `explode` again.

```python
cleaned_titles = cleaned_titles.explode('production_countries')
cleaned_titles.head(10)
```

The structure of `cleaned_titles` looks good now. Let's move on to `cleaned_credits`.

```python
cleaned_credits.head(10)
```

This looks perfect in terms of structure. We don't need to modify anything here.

### Deal with null/missing data

In table `cleaned_titles`, we see that `title`, `description`, `age_certification`, `genres`, `production_countries`, `seasons`, `imdb_id`, `imdb_score`, `tmdb_popularity`, `tmdb_score`, `imdb_votes`, `tmdb_popularity`, `tmdb_score` have null values.

Null values in `title`, `description`, `age_certification`, `production_countries`, `seasons`, `imdb_id`, `tmdb_popularity`, `tmdb_score`, `imdb_votes`, `tmdb_popularity`, `tmdb_score` should not bother us and we will keep them as is.

`imdb_score` and `genres` are closely related to what we are trying to analyze, so let's deep dive into those values.

First, let's look at `imdb_score`.

We want to know great actors from different genres of movies/shows, so let's delete these entries with null values.

```python
cleaned_titles = cleaned_titles.dropna(subset=["genres"])
cleaned_titles["genres"].isnull().sum()
```

In table `cleaned_credits`, we see that `character` has null values.

### Deal with inconsistent data

Sometimes, we might see inconsistent data in countries names. For example, both "US" and "United States" refer to the same country. We will check `production_countries` in table `cleaned_titles`.

```python
with pd.option_context('display.max_rows', None):
    print(cleaned_titles['production_countries'].value_counts())
```

## Retrieve the data that we want

```python
cleaned_titles
```

```python
cleaned_credits
```

To analyze the best actors in each genre, we want to join the tables `cleaned_titles` and `cleaned_credits` by column `id` as it refers to movies/shows in both tables.

```python
credits_with_titles = pd.merge(cleaned_credits, cleaned_titles, on="id", how="inner")
```

```python
credits_with_titles.head()
```

Since we only care about actors, let's filter out the unwanted entries.

```python
actor_with_titles = credits_with_titles.query('role == "ACTOR"')
```

Use `groupby` to group `genres` and `person_id`. Then, we use `mean()` to calculate the average IMDB ratings for an actor in each genre.

```python
groupby_genres_and_person_id = actor_with_titles.groupby(["genres", "person_id"])
```

```python
imdb_score_groupby_genres_and_person_id = groupby_genres_and_person_id["imdb_score"].mean()
imdb_score_groupby_genres_and_person_id
```

We may reset the indice to get a better formatted DataFrame.

```python
imdb_score_groupby_genres_and_person_id_df = imdb_score_groupby_genres_and_person_id.reset_index()
imdb_score_groupby_genres_and_person_id_df
```

Now we have the ratings for every actor in each genre. We also want to find the actor with highest rating in each genre. To do so we first need to find the maximum value in each genre and join the table `imdb_score_groupby_genres_and_person_id_df`.

```python
genres_max_scores = imdb_score_groupby_genres_and_person_id_df.groupby("genres")["imdb_score"].max()
genres_max_scores
```

```python
genres_max_score_with_person_id = pd.merge(imdb_score_groupby_genres_and_person_id_df, genres_max_scores, on=["genres", "imdb_score"])
genres_max_score_with_person_id
```

Now we have what we need. Let's also print out their names.

```python
actor_id_with_names = cleaned_credits[['person_id', 'name']].drop_duplicates()
genres_max_score_with_actor_name = pd.merge(genres_max_score_with_person_id, actor_id_with_names, on="person_id")
genres_max_score_with_actor_name
```

We may sort the table by genres for a better view.

```python
genres_max_score_with_actor_name = genres_max_score_with_actor_name.sort_values("genres").reset_index().drop("index", axis=1)
genres_max_score_with_actor_name.head(20)
```

Now we have reached our goal! Let's store the data as a file.

```python
genres_max_score_with_actor_name.to_csv("genres_max_score_with_actor_name.csv", index=False)
```
